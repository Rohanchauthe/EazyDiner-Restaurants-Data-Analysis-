```python
import pandas as pd 
```

```python
df = pd.read_csv("Eazydiner_data.csv")
df
```

```python
df.dtypes
```

```python
import re

cities = ['Mumbai', 'Pune', 'Delhi','Noida','Gurgaon','saket','vasant kunj','faridabad','parel','Thane','vashi']


pattern = r'\b(' + '|'.join(map(re.escape, cities)) + r')\b'

df['city'] = df['address'].str.extract(pattern, flags=re.IGNORECASE)
```

```python
df
```

```python
df.isna().sum()
```

```python
df['address'] = df['address'].str.split('|').str[0].str.strip()
df
```

```python
df['prices'] = (
    df['prices']
    .astype(str)
    .str.replace(r'\D', '', regex=True)
    .replace('', pd.NA)
    .astype('Int64')
)
```

```python
df
```

```python
cols = list(df.columns)
cols.insert(2, cols.pop(cols.index('city')))  # move to index 2
df = df[cols]
df
```

```python
df.insert(
    df.columns.get_loc('offers') + 1,
    'online payment offer',
    df['offers'].str.split('+').str[-1].str.strip()
)
df
```

```python
df.drop("online payment offer",axis = 1,inplace = True)
df
```

```python
# split offers using +
split_offers = df['offers'].str.split(r'\+', expand=True)

# add second offer to new column
df['online_payment_offer'] = split_offers[1].str.strip()

# keep only first offer in original column
df['offers'] = split_offers[0].str.strip()

df
```

```python
cols = list(df.columns)
cols.insert(7, cols.pop(cols.index('online_payment_offer')))  # move to index 2
df = df[cols]
df
```

```python
cols = list(df.columns)
cols.insert(6, cols.pop(cols.index("Meal_Type")))  # move to index 2
df = df[cols]
df
```

```python
df['online_payment_offer'] = df['online_payment_offer'].fillna('No Offer')
df
```

```python
df.dtypes
```

```python
df = df.rename(columns={'offers': 'Restaurant_offers'})
df
```

```python
df = df.rename(columns={'online_payment_offer': 'Online_payment_offers'})
df
```

```python
df.isna().sum()
```

```python
df['cuisines'] = df['cuisines'].ffill()
df
```

```python
df['Restaurant_offers'] = df['Restaurant_offers'].str.extract(r'(\d+)')
df
```

```python
df['Online_payment_offers'] = df['Online_payment_offers'].str.extract(r'(\d+)')
df
```

```python
df.isna().sum()
```

```python
df['Online_payment_offers'] = df['Online_payment_offers'].fillna('No Offer')
df
```

```python
df['Restaurant_offers'] = df['Restaurant_offers'].fillna('No Offer')
df
```

```python
df.isna().sum()
```

```python
df['ratings'] = df['ratings'].fillna(df['ratings'].mean())
df
```

```python
df = df.rename(columns={'Online_payment_offers': 'Online_payment_offer_percentage'})
df
```
